```python
print("Simple Calculator")
val1 = float(input("Enter first value: "))
val2 = float(input("Enter second value: "))
operator = input("Enter operator (+, -, *, /, **): ")
if operator == '+':
    print(val1 + val2)
elif operator == '-':
    print(val1 - val2)
elif operator == '*':
    print(val1 * val2)
elif operator == '/':
    print(val1 / val2)
elif operator == '**':
    print(val1 ** val2)
else:
    print("Invalid operator!")
```

```python
print("Check for Numeric Value in a List")
mixed_list = ["Apple", "Orange", 42, "Banana", 7.5]

for item in mixed_list:
    if isinstance(item, (int, float)):
        print(f"Numeric value found: {item}")
```

```python
print("Add a Key to a Dictionary")
customer = {"first_name": "Ali", "last_name": "Khan"}
# Adding a new key 'address'
customer["address"] = "Karachi, Pakistan"

print(customer)
```

```python
print("Sum All Numeric Items in a Dictionary")
data_dict = {"a": 100, "b": "Text", "c": 250, "d": 50}
total_sum = 0

for val in data_dict.values():
    if isinstance(val, (int, float)):
        total_sum += val

print("Sum of numeric items:", total_sum)
```

```python
print("Identify Duplicate Values from a List")
my_list = [1, 2, 3, 2, 4, 5, 1, 6, 7, 5]
duplicates = []
unique_items = []

for x in my_list:
    if x in unique_items:
        if x not in duplicates:
            duplicates.append(x)
    else:
        unique_items.append(x)

print("Duplicate values are:", duplicates)
```

```python
inventory = {"apples": 10, "bananas": 5, "oranges": 8}
key_to_check = "bananas"

if key_to_check in inventory:
    print(f"Key '{key_to_check}' exists! Value: {inventory[key_to_check]}")
else:
    print(f"Key '{key_to_check}' does not exist.")
```
